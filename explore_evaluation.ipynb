```python
import pandas as pd

filename = "results/test-extrinsic-100"
# filename = "results/test-xent-100"

df_aggregated = pd.read_csv(f"{filename}.csv").sort_values("factual", ascending=False).set_index("model")
df_sums = pd.read_json(f"{filename}-summaries.json")

df_aggregated.loc["fbs_oracle", "skipped"] += df_aggregated.loc["fbs_oracle"]["non_factual"]
df_aggregated.loc["fbs_oracle", "non_factual"] = 0

df_aggregated
```

```python
import pandas as pd

df_full = pd.read_csv("results/full.csv").sort_values("factual", ascending=False).set_index("model")
df_full_extrinsic = pd.read_csv("results/full-extrinsic.csv").sort_values("factual", ascending=False).set_index("model")
df_full
```

```python
df_full_extrinsic
```

```python
# Export to latex
model_mapping = [
    ("fbs_oracle", "fbsOracle"),
    ("fbs_classifier", "fbsClassifier"),
    ("baseline", "baseline"),
    ("corrector", "corrector"),
]
label_mapping = [
    ("factual", "Factual"),
    ("non_factual", "NonFactual"),
    ("non_factual_extrinsic", "NonFactualExtrinsic"),
    ("non_factual_intrinsic", "NonFactualIntrinsic"),
    ("skipped", "Skipped"),
    ("rouge1", "RougeOne"),
    ("rouge2", "RougeTwo"),
    ("rougeL", "RougeL")
]
with open("results/latex.txt", "w") as f:
    for model_index, model_label in model_mapping:
        for metric_index, metric_label in label_mapping:
            if "rouge" in metric_index and model_index in ["fbs_classifier", "baseline", "corrector"]:
                metric = df_full.loc[model_index][metric_index]
                str = f"\\newcommand{{\\{model_label}{metric_label}Full}}{{{metric*100:.2f}}}" + "\n"
                f.write(str.replace("%", "\\%"))
                metric = df_full_extrinsic.loc[model_index][metric_index]
                str = f"\\newcommand{{\\{model_label}{metric_label}Extrinsic}}{{{metric*100:.2f}}}" + "\n"
                f.write(str.replace("%", "\\%"))
            elif "rouge" not in metric_index:
                metric = df_aggregated.loc[model_index][metric_index]
                str = f"\\newcommand{{\\{model_label}{metric_label}}}{{{metric:.0%}}}" + "\n"
                f.write(str.replace("%", "\\%"))
```

```python
# Helper method
def print_summaries(df, models):
    for sum_id in df.index:
        print(f"---XSUM ID {sum_id}---")
        row = df.loc[sum_id]
        for model in models:
            n_ents = row[model + '_n_extrinsic']
            summary = row[model + '_summary']
            print(f"{model} ({n_ents} extrinsic ents): {[summary]}")
        print()
```

### Summaries corrected by Oracle

```python
df_inspect = df_sums[
    df_sums.fbs_oracle_is_factual
    & df_sums.baseline_is_non_factual
]
print(df_inspect.shape[0])
print_summaries(
    df_inspect,
    [
        "baseline",
        "fbs_oracle"
    ]
)
```

### Summaries messed by FBS Classifier

```python
df_inspect = df_sums[
    df_sums.baseline_is_factual
    & df_sums.fbs_classifier_is_non_factual
]

print(df_inspect.shape)
print_summaries(
    df_inspect,
    [
        "baseline",
        "fbs_classifier"
    ]
)
```

### Corrected by FBS Classifier

```python
df_inspect = df_sums[
    df_sums.baseline_is_non_factual
    & df_sums.fbs_classifier_is_factual
]

print(df_inspect.shape)
print_summaries(
    df_inspect,
    [
        "baseline",
        "fbs_classifier"
    ]
)
```

### Summaries messed up by Corrector

```python
df_inspect = df_sums[
    df_sums.baseline_is_factual
    & df_sums.corrector_is_non_factual
]
print(df_inspect.shape[0])
print_summaries(
    df_inspect,
    [
        "baseline",
        "corrector"
    ]
)
```

```python
df_inspect = df_sums[
    df_sums.fbs_oracle_is_factual
    & df_sums.corrector_is_non_factual
]
print(df_inspect.shape[0])
print_summaries(
    df_inspect,
    [
        "fbs_oracle",
        "corrector"
    ]
)
```
